```python
import numpy as np
# 파이썬에서 수치 계산을 위해서 사용
import pandas as pd
# 파이썬에서 데이터 분석 위해 사용
from sklearn.datasets import load_iris
# sklearn : 머신러닝 특화된 모듈

# 데이터 불러오기
data = load_iris()

data
```

```python
# 독립변수
x = pd.DataFrame(data.data, columns=data.feature_names)
# 종속변수
y = pd.DataFrame(data.target, columns=['Species'])

# concatenate 결합하다
pd.concat([x,y], axis=1)

# axis=0 : 서로 다른 행끼리, 열기준
# axis=1 : 서로 다른 열끼리, 행기준
# 행세 열가
```

```python
# 지도학습(분류) 예 - 샘플코드 1

from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
```

```python
data.data
```

```python
x = data.data
y = data.target
```

```python
x
```

```python
y
```

```python
x[:, :10] # 데이터 속성 일부만 가져오기
```

```python
# 구현

from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver = 'lbfgs')
```

```python
model.fit(x,y)
```

```python
# 예측
y_pred = model.predict(x)
```

```python
# 우리의 목표
# 예측값인 y_pred(y_hat)와 실제값인 y(y_real) 차이 >> 오차
# >> 오차를 줄이는 것이 목표이다.

from sklearn.metrics import accuracy_score

accuracy_score(y, y_pred)
```

```python
# 비지도 학습의 특징 : 정답(라벨)이 없다

from sklearn.datasets import load_wine

data = load_wine()
```

```python
data
```

```python
data.data[:, [0,9]] # 0, 9 두가지 데이터만 가져옴
```

```python
x = data.data[:, [0,9]]
```

```python
from sklearn.cluster import KMeans

n_cluster = 3
```

```python
model = KMeans(n_clusters = n_cluster)
```

```python
pred = model.fit_predict(x) # 비지도 학습은 답(라벨)이 없음 >> y가 없음.
# >> x 를 기준 (여기서는 3집단)으로 구분하는 예측을 실시.
```

```python
# 데이터 시각화

%matplotlib inline
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

# 마커 's' 사각형 square 의 약자,
ax.scatter(x[pred==0, 0], x[pred==0,1], color='red', marker='s', label='label1' )
ax.scatter(x[pred==1, 0], x[pred==1,1], color='blue', marker='s', label='label2' )
ax.scatter(x[pred==2, 0], x[pred==2,1], color='green', marker='s', label='label3' )
ax.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], s=200,
           color='yellow', marker='*', label='center') # s=size

ax.legend() #범례
plt.show()
```
